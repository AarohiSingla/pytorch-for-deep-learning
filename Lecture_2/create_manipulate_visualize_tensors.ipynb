# 1D tensor

In this example, we create a 1D tensor (tensor) with some initial data, print the original tensor, square each element to manipulate it, print the manipulated tensor, and then visualize the tensor using a simple line plot. The numpy() method is used to convert the PyTorch tensor to a NumPy array for plotting.

```python
import torch

# Create a 1D tensor
tensor_data = [1, 2, 3, 4, 5]
tensor = torch.tensor(tensor_data, dtype=torch.float32)

# Display the created tensor
print("Original Tensor:")
print(tensor)
```

```python
# Manipulate the tensor
# Example: Square each element
tensor = tensor ** 2

# Display the manipulated tensor
print("Manipulated Tensor (Squared):")
print(tensor)
```

## 2D tensor

```python
import torch


# Create a tensor
tensor_data = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
tensor = torch.tensor(tensor_data, dtype=torch.float32)


# Display the created tensor
print("Original Tensor:")
print(tensor)
```

```python
# Manipulate the tensor
# Example: Multiply all elements by 2
tensor *= 2

# Display the manipulated tensor
print("\nManipulated Tensor (Multiplied by 2):")
print(tensor)
```

```python
import matplotlib.pyplot as plt

# Visualize the tensor
plt.imshow(tensor, cmap='viridis', interpolation='nearest')
plt.title("Visualization of Manipulated Tensor")
plt.colorbar()
plt.show()
```

We visualize the manipulated tensor using matplotlib.pyplot.imshow() to create a simple heatmap.
The cmap parameter specifies the color map to use; in this case, 'viridis' is a perceptually uniform color map. The values in the tensor are mapped to colors according to the color map.

```python
import torch
import matplotlib.pyplot as plt
from torchvision import transforms
from PIL import Image 
 
# Load an image using PIL (Python Imaging Library)
image_path = "flower.jpg"  # Replace with the actual path to your image
image = Image.open(image_path)

# Display the original image
plt.imshow(image)
plt.title("Original Image")
plt.axis('off')
plt.show()

# Convert the PIL image to a PyTorch tensor
image_transform = transforms.ToTensor()
tensor_image = image_transform(image)

# Display the PyTorch tensor version of the image
plt.imshow(tensor_image.permute(1, 2, 0))  # permute to change channel order
plt.title("Tensor Version of the Image")
plt.axis('off')
plt.show()
```

In this example:

Load an image using the Python Imaging Library (PIL). You need to replace "path/to/your/image.jpg" with the actual path to your image file.

Display the original image using matplotlib.pyplot.

Convert the PIL image to a PyTorch tensor using transforms.ToTensor() from the torchvision module.

Display the PyTorch tensor version of the image. Note the use of permute to change the channel order since PyTorch uses a different channel order (C, H, W) compared to the default order in PIL (H, W, C).

A colored image is typically represented as a 3D tensor. The three dimensions correspond to the color channels: red, green, and blue (RGB). Each channel contains pixel values representing the intensity of the respective color at each pixel in the image.

For example, a colored image with dimensions 256x256 pixels would be represented as a 3D tensor with shape (3, 256, 256). 

It's important to note that a grayscale image, which has only one channel representing intensity, is represented as a 2D tensor with shape (H, W).
For example, a grayscale image with dimensions 256x256 pixels would be represented as a 2D tensor with shape (256, 256). 

```python
import torch
import matplotlib.pyplot as plt
from torchvision import transforms
from PIL import Image

# Load an image using PIL (Python Imaging Library)
image_path = "flower.jpg"  # Replace with the actual path to your image
image = Image.open(image_path)

# Display the original image
# plt.imshow(image)
# plt.title("Original Image")
# plt.axis('off')
# plt.show()

# Convert the PIL image to a PyTorch tensor
image_transform = transforms.ToTensor()
tensor_image = image_transform(image)

# Display the PyTorch tensor version of the image
# plt.imshow(tensor_image.permute(1, 2, 0))  # permute to change channel order
# plt.title("Tensor Version of the Image")
# plt.axis('off')
# plt.show()

# Print the PyTorch tensor
print("PyTorch Tensor:")
print(tensor_image)
```

By adding the print(tensor_image) statement at the end, you will see the printed representation of the PyTorch tensor, which includes the pixel values for each channel. The tensor will be displayed in the format (channel, height, width) with the actual pixel values.

Each value in the tensor represents the intensity of the corresponding pixel in the image. The values are normalized between 0 and 1.
In your specific tensor, the pixel values seem to be very close, indicating that the image is relatively dark. If the pixel values were closer to 1, it would suggest a brighter image.

To better understand the pixel values, you can look at specific values within the tensor. For example, to access the pixel value at row 0, column 0, and the first channel, you can use tensor[0, 0, 0]

```python
tensor_image[0, 0, 0]
```

The output tensor(0.1804) indicates that the pixel value at the first channel, first row, and first column of your image is 0.1804

```python
tensor_image[1, 0, 0]
```

```python
tensor_image[2, 0, 0]
```

The outputs you're seeing, such as tensor(0.1804), tensor(0.2784), and tensor(0.1137), represent the normalized pixel values at specific locations in your image tensor.

In PyTorch, tensors are used to represent numerical data,

    tensor_image[0, 0, 0] gives tensor(0.1804): This means that the red channel value at the top-left corner of your image is approximately 0.1804.
    
    tensor_image[1, 0, 0] gives tensor(0.2784): This corresponds to the green channel value at the same pixel location, which is approximately 0.2784.
    
    tensor_image[2, 0, 0] gives tensor(0.1137): This represents the blue channel value at the top-left corner, which is approximately 0.1137.

    tensor_image[0, 0, 0] (approximately 0.1804) represents the intensity of red in that pixel.
    
    tensor_image[1, 0, 0] (approximately 0.2784) represents the intensity of green.
    
    tensor_image[2, 0, 0] (approximately 0.1137) represents the intensity of blue.

Together, these three values define the color of the pixel in terms of its RGB components. The overall color of the pixel is a combination of these three intensities. In this specific case, the pixel color at the top-left corner of your image is a shade of color defined by the specific values in the red, green, and blue channels. 

## 2D Image

```python
import torch
import matplotlib.pyplot as plt
from torchvision import transforms
from PIL import Image

# Load a grayscale image using PIL (Python Imaging Library)
image_path = "grey.jpg"  
image = Image.open(image_path)

# Display the original image
plt.imshow(image, cmap='gray')  # Use 'gray' colormap for grayscale images
plt.title("Original Image")
plt.axis('off')
plt.show()

# Convert the PIL image to a PyTorch tensor
image_transform = transforms.ToTensor()
tensor_image = image_transform(image)

# Display the PyTorch tensor version of the image
plt.imshow(tensor_image[0], cmap='gray')  # Access the single channel
plt.title("Tensor Version of the Image")
plt.axis('off')
plt.show()

# Print the PyTorch tensor
print("PyTorch Tensor:")
print(tensor_image)
```

In PyTorch, when you use transforms.ToTensor() on a grayscale image, it still converts the image to a 3D tensor. The resulting tensor has a single channel (dimension), but PyTorch still represents it as a 3D tensor to maintain consistency in handling various types of images.

Although the grayscale image doesn't have multiple color channels, PyTorch represents it as a 3D tensor for consistency in handling various types of images. In a grayscale image tensor, the single channel represents the intensity of the grayscale values. Accessing tensor_image[0] gives you the 2D tensor of the grayscale image.

If you print the tensor, you'll see that it has a shape like (1, height, width), indicating a single channel.
